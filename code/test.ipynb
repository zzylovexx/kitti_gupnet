```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
from lib.datasets.kitti_utils import get_affine_transform
```

```python
img_file='Dataset/KITTI/training/image_2/000004.png'
img=Image.open(img_file)
scale=0.5
shift=0.1
display(img)
img_size = np.array(img.size)
resolution=np.array([1280,384])
random_crop_flag = True
center = np.array(img_size) / 2
crop_size = img_size * np.clip(np.random.randn()*scale + 1, 1 - scale, 1 + scale)
center[0] += img_size[0] * np.clip(np.random.randn() * shift, -2 * shift, 2 * shift)
center[1] += img_size[1] * np.clip(np.random.randn() * shift, -2 * shift, 2 * shift)
# display(img)
# print(img_size)


crop_size = img_size
# img = img.transpose(Image.FLIP_LEFT_RIGHT)
# display(img)

trans, trans_inv = get_affine_transform(center, crop_size, 0, resolution, inv=1)
img = img.transform(tuple(resolution.tolist()),
                            method=Image.AFFINE,
                            data=tuple(trans_inv.reshape(-1).tolist()),
                            resample=Image.BILINEAR)
display(img)
print(img.size)
```
